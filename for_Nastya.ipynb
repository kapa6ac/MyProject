```python
from bs4 import BeautifulSoup
import re
import requests
import time
import json
import lxml
import urllib3
import pandas as pd
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
def glava_filling(u):
    return ['https://constrf.ru' + u.attrs.get('href'), u.text]
request = requests.get('https://constrf.ru/razdel-1/glava-1', verify = False)
```

```python
bs = BeautifulSoup(request.text,"lxml").find_all('a')
df = pd.DataFrame(columns=['url', 'title'])

for u in bs:
    if 'Статья' in u.text:
        df.loc[len(df['url'])] = glava_filling(u)
df = df.drop_duplicates(subset='url')
```

```python
def parsing_func(u):
    t=""
    request = requests.get(u, verify = False)
    for i in BeautifulSoup(request.text,"lxml").find_all('p'):
        t += i.get_text()
    return t
```

```python
df['text'] = [parsing_func(u) for u in df['url']]
```
